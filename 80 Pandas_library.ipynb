## Data Cleaning

### Handling Missing Data

```python
import pandas as pd
import numpy as np

data = {'A': [1, 2, np.nan, 4, 5],
        'B': [np.nan, 2, 3, 4, 5]}

df = pd.DataFrame(data)

# Fill missing values with a specific value (e.g., 0)
df.fillna(0, inplace=True)

df
```

### Data Imputation

```python
import pandas as pd
import numpy as np

data = {'A': [1, 2, np.nan, 4, 5],
        'B': [np.nan, 2, 3, 4, 5]}

df = pd.DataFrame(data)

# Fill missing values with the mean of each column
df.fillna(df.mean(), inplace=True)

df
```

### Data Transformation

```python
import pandas as pd

data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Square the values in column 'A'
df['A'] = df['A'].apply(lambda x: x**2)
df
```

### Removing Duplicates

```python
import pandas as pd

data = {'A': [1, 2, 2, 4, 5],
        'B': ['apple', 'banana', 'banana', 'cherry', 'date']}

df = pd.DataFrame(data)

# Remove duplicate rows based on column 'A'
df.drop_duplicates(subset='A', keep='first', inplace=True)
df
```

### Data **Filtering**

```python
import pandas as pd

data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Filter rows where column 'A' is greater than 3
filtered_df = df[df['A'] > 3]

filtered_df
```

### Filtering Data

### Using Multiple Conditions

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}

df = pd.DataFrame(data)

# Filter rows where Age is greater than 25 and City is 'San Francisco'
condition = (df['Age'] > 25) & (df['City'] == 'San Francisco')
filtered_data = df[condition]
filtered_data
```

## Data Aggregation

### Grouping Data


```python
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Group data by the 'Category' column
grouped = df.groupby('Category')
grouped
```

### Aggregating Data

```python
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Group data by the 'Category' column and calculate the sum for each group
grouped = df.groupby('Category')
result = grouped['Value'].sum()
result
```

### Multiple Aggregation Functions

```python
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Group data by the 'Category' column and calculate sum and mean for each group
grouped = df.groupby('Category')
result = grouped['Value'].agg(['sum', 'mean'])
result
```

### Custom Aggregation Functions

```python
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Define a custom aggregation function
def custom_agg(x):
    return x.max() - x.min()

# Group data by the 'Category' column and apply the custom aggregation function
grouped = df.groupby('Category')
result = grouped['Value'].agg(custom_agg)
result
```

## Integration with NumPy

```python
import pandas as pd
import numpy as np

# Create a Pandas Series and a NumPy array
data = {'A': [1, 2, 3, 4, 5]}
series = pd.Series(data['A'])
numpy_array = np.array([10, 20, 30, 40, 50])

# Perform operations on the Series and NumPy array
result1 = series + numpy_array
result2 = series * 2

print(result1)
print(result2)
```

```python
import pandas as pd

data = {'A': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Convert a Pandas DataFrame to a NumPy array
numpy_array = df.values  # or df.to_numpy()

# Convert a NumPy array to a Pandas DataFrame
new_df = pd.DataFrame(numpy_array, columns=['A'])
new_df
```

## Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Sales': [100, 120, 130, 110, 150]}

df = pd.DataFrame(data)
df
```

```python
# Create a line plot
plt.plot(df['Month'], df['Sales'])
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
# Select and filter data
high_sales_months = df[df['Sales'] > 120]

# Create a bar plot for high-sales months
plt.bar(high_sales_months['Month'], high_sales_months['Sales'], color='royalblue')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('High Sales Months')
plt.show()
```
